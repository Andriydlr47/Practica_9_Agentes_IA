```python
import pandas as pd
import reverse_geocoder as rg
```

```python
df = pd.read_csv("RAG/vias_espania_8anu.csv")
```

### Limpiamos los nulos

```python
df.dropna(subset=['lat', 'lon'], inplace=True)
```

### Preparamos las tuplas de coordenadas

```python
coords = tuple(zip(df['lat'], df['lon']))
```

### Obtenemos los resultados del geodecoder inverso

```python
results = rg.search(coords)
```

```python
df['ciudad'] = [res.get('name', '') for res in results]
df['comunidad_autonoma'] = [res.get('admin1', '') for res in results]
```

```python
df.to_csv("RAG/vias_espania_8anu_limpio.csv", index=False)
print("Dataset actualizado y guardado correctamente con Ciudad y CCAA.")
```
